```python
import os
import numpy as np
from shutil import copyfile
```

```python
def LoadAllCif(Path):
    CifInfo = []
    Path_i = sorted(os.listdir(Path))
    for i in Path_i:
        print(i)
        with open(Path + "/" + i) as textFile:
            lines = [line.split() for line in textFile]

        for j in lines:
            for k in range(0, len(j)):
                if(j[k] == "_cell_length_a"):
                    lattice_a = float(j[k + 1].split("(")[0])
                    print(lattice_a)
                
                if(j[k] == "_symmetry_Int_Tables_number"):
                    spacegroup = j[k+1]
                    print(spacegroup)
                    print("")
                    CifInfo.append([int(i.split(".")[0]), lattice_a, spacegroup])
                    break
    return(CifInfo)
```

```python
Path = "/home/ug-ml/felix-ML/VAE_000/DataAllInOne_Normalised/VAE_000_2/CifFlolder"

CifInfo = LoadAllCif(Path)
```

```python
def IsInList(List, Value):
    if(len(List) == 0):
        return(False)
    else:
        for i in List:
            if(i == Value):
                return True
    return False

def SpaceGroupInfo(CifInfo):
    ListSpaceGroup = []
    for i in CifInfo:
        spacegroup = i[2]
        if(IsInList(ListSpaceGroup, spacegroup) == False):
            ListSpaceGroup.append(spacegroup)
    ListSpaceGroup = sorted(ListSpaceGroup)
    NumberSpaceGroup = len(ListSpaceGroup)
    HistSpaceGroup = np.zeros(NumberSpaceGroup, dtype = np.int)
    
    for i in CifInfo:
        spacegroup = i[2]
        arg = 0
        for j in range(0, NumberSpaceGroup):
            if(ListSpaceGroup[j] == spacegroup):
                arg = j
                break
        HistSpaceGroup[j]+=1
        
    
    return(ListSpaceGroup, HistSpaceGroup)
```

```python
ListSpaceGroup, HistSpaceGroup = SpaceGroupInfo(CifInfo)

print(ListSpaceGroup)
print(HistSpaceGroup)

print(np.shape(HistSpaceGroup))
print(np.sum(HistSpaceGroup))

sum = 0
for i in HistSpaceGroup:
    if(i >= 100):
        sum+= i
    
print(sum)
```

```python
def LatticeSizeInfo(CifInfo):
    minsize = np.inf
    maxsize = -np.inf
    for i in CifInfo:
        if(i[1] < minsize):
            minsize = i[1]
        elif(i[1] > maxsize):
            maxsize = i[1]
            
    NumberLattice = int(np.floor(maxsize) - np.floor(minsize) + 1)
    LatticeHist = np.zeros(NumberLattice, dtype = np.int)
    for i in CifInfo:
        index = int(np.floor(i[1]) - np.floor(minsize))
        LatticeHist[index]+=1
    
    return([minsize, maxsize], LatticeHist)
```

```python
LatticeRange, LatticeHist = LatticeSizeInfo(CifInfo)
```

```python
print(LatticeRange)
print(LatticeHist)
StartLatticeIndex = 1
EndLatticeIndex = 14

print(np.sum(LatticeHist[StartLatticeIndex:EndLatticeIndex]))
```

```python
#Want to only keep space group if there are larger than a certain number of crystals
#Want to keep a reduced range of lattice parameters, some values values are too uncommon

#SG = SpaceGroup
KeepNumberSG = 100
KeepSG = []

for i in range(len(HistSpaceGroup)):
    if(HistSpaceGroup[i] >= KeepNumberSG):
        KeepSG.append(ListSpaceGroup[i])


KeepRangeOfLattice = [int(np.floor(LatticeRange[0]) + StartLatticeIndex), int(np.floor(LatticeRange[0]) + EndLatticeIndex)]
print(KeepRangeOfLattice)
print(KeepSG)
```

```python
def RemoveCrystalNo(CifInfo, KeepRangeOfLattice, KeepSG):
    RemoveCrystals = []
    for i in CifInfo:
        if(i[1] < KeepRangeOfLattice[0] or i[1] > KeepRangeOfLattice[1]):
            RemoveCrystals.append(i[0])
        elif(IsInList(KeepSG, i[2]) == False):
            RemoveCrystals.append(i[0])
    return(RemoveCrystals)
```

```python
RemoveCrystals = RemoveCrystalNo(CifInfo, KeepRangeOfLattice, KeepSG)
RemoveCrystals = np.array(RemoveCrystals)
print(len(RemoveCrystals))
print(RemoveCrystals)

np.save("RemoveCrystals.npy", RemoveCrystals)
```
