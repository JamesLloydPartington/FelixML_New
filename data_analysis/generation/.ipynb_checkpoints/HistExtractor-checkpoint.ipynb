```python
import numpy as np
import matplotlib.pyplot as plt
import os
import math


path = "/home/ug-ml/Downloads/VAE_Hist"


```

```python
with open(path) as textFile:
    lines = [line.split() for line in textFile]

    
training_loss = []
validation_loss = []
for i in lines:
    if(i[0] == "139/139"):
        for j in range(0, len(i)):
            if(i[j] == "reconstruction_loss:"):
                training_loss.append(float(i[j+1]))
                break
    if(i[0] == "Average"):
        for j in range(0, len(i)):
            if(i[j] == "loss:"):
                validation_loss.append(float(i[j+1]))
                break
```

```python
plt.plot(np.arange(0, len(training_loss)), training_loss, "o", label = "Training")
plt.plot(np.arange(0, len(validation_loss)), validation_loss, "o", label = "Validation")


plt.xscale("log")
plt.yscale("log")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Title')
plt.legend()
plt.grid()
plt.show()


save_vae = "/home/ug-ml/tmp_git/FelixML_New/data_analysis/generation/Graphs/Epochs_Loss/"
np.savetxt(save_vae + "training_loss_vae", training_loss)
np.savetxt(save_vae + "validation_loss_vae", validation_loss)
```

```python
path_cnn = "/home/ug-ml/Downloads/CNN_Hist"

with open(path_cnn) as textFile:
    lines_cnn = [line.split() for line in textFile]

    
training_loss_cnn = []
validation_loss_cnn = []

training_acc_cnn = []
validation_acc_cnn = []


for i in lines_cnn:
    if(i[0] == "1769/1769"):
        for j in range(0, len(i)):
            if(i[j] == "loss:"):
                if(i[j+1][-1] == ":"):
                    training_loss_cnn.append(float(i[j+1][0:-1]))
                else:
                    training_loss_cnn.append(float(i[j+1]))
            if(i[j] == "acc:"):
                if(i[j+1][-1] == ":"):
                    training_acc_cnn.append(float(i[j+1][0:-1]))
                else:
                    training_acc_cnn.append(float(i[j+1]))
                break
    
    if(i[0] == "208/208"):
        for j in range(0, len(i)):
            if(i[j] == "loss:"):
                #print(i)
                if(i[j+1][-1] == ":"): 
                    validation_loss_cnn.append(float(i[j+1][0:-1]))
                else:
                    validation_loss_cnn.append(float(i[j+1]))
            if(i[j] == "acc:"):
                if(i[j+1][-1] == ":"):
                    validation_acc_cnn.append(float(i[j+1][0:-1]))
                else:
                    validation_acc_cnn.append(float(i[j+1]))
                break
```

```python


plt.plot(np.arange(0, len(training_acc_cnn)), training_acc_cnn, "o", label = "Training")
plt.plot(np.arange(0, len(validation_acc_cnn)), validation_acc_cnn, "o", label = "Validation")


#plt.xscale("log")
#plt.yscale("log")
plt.xlabel('Epochs')
plt.ylabel('Acc')
plt.title('Title')
plt.legend()
plt.grid()
plt.show()


plt.plot(np.arange(0, len(training_loss_cnn)), training_loss_cnn, "o", label = "Training")
plt.plot(np.arange(0, len(validation_loss_cnn)), validation_loss_cnn, "o", label = "Validation")


plt.xscale("log")
plt.yscale("log")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Title')
plt.legend()
plt.grid()
plt.show()


save_cnn = "/home/ug-ml/tmp_git/FelixML_New/data_analysis/classification/Graphs/Epochs_LossAcc/"
np.savetxt(save_cnn + "training_acc_cnn", training_acc_cnn)
np.savetxt(save_cnn + "validation_acc_cnn", validation_acc_cnn)
np.savetxt(save_cnn + "training_loss_cnn", training_loss_cnn)
np.savetxt(save_cnn + "validation_loss_cnn", validation_loss_cnn)
```
